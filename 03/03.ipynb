```python
from pathlib import Path
from collections import defaultdict

def load_matrix(path: str):
    mat = []
    with Path(path).open() as f:
        for line in f.readlines():
            mat.append(["."] + [c for c in line if c != "\n"] + ["."])
    n = len(mat[0])
    return [["."] * n] + mat + [["."] * n]
```

```python
mat = load_matrix("./03_test.txt")
buffer = ""
nums = []
m = len(mat)
n = len(mat[0])

for y in range(1, m):
    for x in range(1, n):
        if mat[y][x].isdigit():
            buffer += mat[y][x]
        elif buffer:
            num = int(buffer)
            for i in range(x-len(buffer)-1, x+1):
                if ((not mat[y-1][i].isdigit() and mat[y-1][i] != ".") or
                    (not mat[y+1][i].isdigit() and mat[y+1][i] != ".")):
                    nums.append(num)
                    break
            else:
                if ((not mat[y][x].isdigit() and mat[y][x] != ".") or
                    (not mat[y][x-len(buffer)-1].isdigit() and mat[y][x-len(buffer)-1] != ".")):
                    nums.append(num)
            buffer = ""
assert sum(nums) == 4361
```

```python
gears = defaultdict(list)
buffer = ""
for y in range(1, m):
    for x in range(1, n):
        if mat[y][x].isdigit():
            buffer += mat[y][x]
        elif buffer:
            num = int(buffer)
            for i in range(x-len(buffer)-1, x+1):
                for j in [y-1, y+1]:
                    if mat[j][i] == "*":
                        gears[(j, i)].append(num)
            for i in [x-len(buffer)-1, x]:
                if mat[y][i] == "*":
                    gears[(y, i)].append(num)
            buffer = ""

res = 0
for gear in gears.values():
    if len(gear) == 2:
        res += gear[0] * gear[1]
assert res == 467835
```
